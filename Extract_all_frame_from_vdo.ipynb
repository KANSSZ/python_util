### This code use for extract PIC from VDO and remove duplicate(100%) pic.

Separate code below

```python
import os
import cv2
import imagehash
from PIL import Image

# Path to your video file
video_path = 'KANSSZ.mp4'

# Directory to save unique frames
output_dir = 'KANSSZ'

# Create the directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Open the video file
cap = cv2.VideoCapture(video_path)

frame_count = 0
hash_dict = {}  # Dictionary to store image hashes and their file paths

# Loop over frames in the video
while cap.isOpened():
    ret, frame = cap.read()  # Read the next frame
    if not ret:
        break  # Break the loop when the video ends

    # Save each frame as a temporary image file for hash checking
    temp_filename = os.path.join(output_dir, f'temp_frame_{frame_count:05d}.jpg')  
    cv2.imwrite(temp_filename, frame)

    # Open the saved frame and calculate its hash
    try:
        with Image.open(temp_filename) as img:
            img_hash = imagehash.average_hash(img)

            # Check if the hash already exists in the dictionary
            if img_hash in hash_dict:
                print(f"Duplicate frame found: {temp_filename}. Deleting duplicate.")
                os.remove(temp_filename)  # Delete the duplicate frame
            else:
                # Move the frame from temp to final saved name since it's unique
                final_filename = os.path.join(output_dir, f'frame_{frame_count:05d}.jpg')
                os.rename(temp_filename, final_filename)  # Rename temp to final frame name
                hash_dict[img_hash] = final_filename  # Store the hash with the frame name

        frame_count += 1

    except Exception as e:
        print(f"Error processing frame {frame_count}: {e}")
        if os.path.exists(temp_filename):
            os.remove(temp_filename)  # Cleanup the temp file if error occurred

# Release the video capture object
cap.release()

print(f"Extracted and saved {len(hash_dict)} unique frames to {output_dir}")
```
